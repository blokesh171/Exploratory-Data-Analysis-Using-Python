# Cleaning the normalized losses Column

```python
auto_df['normalized-losses'][auto_df['normalized-losses'] == '?'].count()#count ? values
```

```python
# Setting the missing value to mean of normalized losses and conver the datatype to integer
normalized_loss = auto_df['normalized-losses'].loc[auto_df['normalized-losses'] != '?']
nlmean = normalized_loss.astype(str).astype(int).mean()
auto_df['normalized-losses'] = auto_df['normalized-losses'].replace('?',nlmean).astype(int)
auto_df['normalized-losses'].head()
```

# Cleaning the price column

```python
auto_df['price'].str.isnumeric().value_counts()# all string char are numerical -check
```

```python
auto_df['price'][auto_df['price'].str.isnumeric() == False]
```

```python
price = auto_df['price'][auto_df['price'] != '?']
pmean = price.astype(str).astype(int).mean()
auto_df['price'] = auto_df['price'].replace('?',pmean).astype(int)
auto_df['price'].tail()
```

# Cleaning the horsepower

```python
auto_df['horsepower'][auto_df['horsepower'].str.isnumeric().value_counts()]
```

```python
horsepower = auto_df['horsepower'][auto_df['horsepower'] != '?']
hpmean = horsepower.astype(str).astype(int).mean()
auto_df['horsepower'] = auto_df['horsepower'].replace('?',hpmean).astype(int)
auto_df['horsepower'].head()
```

# Cleaning bore column

```python
auto_df['bore'] = pd.to_numeric(auto_df['bore'],errors='coerce')
auto_df.dtypes
```

# Cleaning the stroke column

```python
auto_df['stroke'] = pd.to_numeric(auto_df['stroke'],errors='coerce')
auto_df.dtypes
```
