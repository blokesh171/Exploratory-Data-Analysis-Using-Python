# univariate analysis


```python
#READ DATA

import pandas as pd
import numpy as np
auto_df=pd.read_excel('C:/Users/tech2k8/data clean nxt to data wrangling.xlsx','auto')
del auto_df['Unnamed: 0']
auto_df
```

# import matplotlib libraries


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Vehicle make frequency diagram

```python
auto_df.make.value_counts().plot(kind='bar', figsize=(15,5))
plt.title("Number of vehicles by make")
plt.ylabel('Number of vehicles')
plt.xlabel('Make');
```

# Insurance risk ratings Histogram

```python
auto_df.symboling.hist(bins=6,color='green')
plt.title("Insurance risk ratings of vehicles")
plt.ylabel('Number of vehicles')
plt.xlabel('Risk rating')
```

# Normalized losses histogram

```python
auto_df['normalized-losses'].hist(bins=5,color='orange');
plt.title("Normalized losses of vehicles")
plt.ylabel('Number of vehicles')
plt.xlabel('Normalized losses')
```

# Fuel type bar chart

```python
auto_df['fuel-type'].value_counts().plot(kind='bar',color='purple')
plt.title("Fuel type frequence diagram")
plt.ylabel('Number of vehicles')
plt.xlabel('Fuel type')
```

# Fuel type pie diagram

```python
auto_df['aspiration'].value_counts().plot.pie(figsize=(6, 6),labels=['gas','dissel'], autopct='%.2f')
plt.title("Fuel type pie diagram")
plt.ylabel('Number of vehicles')
plt.xlabel('Fuel type')
```

# Horse power histogram

```python
auto_df.horsepower.hist(bins=5,color='red');
plt.title("Horse power histogram")
plt.ylabel('Number of vehicles')
plt.xlabel('Horse power')
```

# Curb weight histogram

```python
auto_df['curb-weight'].hist(bins=5,color='brown');
plt.title("Curb weight histogram")
plt.ylabel('Number of vehicles')
plt.xlabel('Curb weight')
```

# Drive wheels bar chart

```python
auto_df['drive-wheels'].value_counts().plot(kind='bar',color='grey')
plt.title("Drive wheels diagram")
plt.ylabel('Number of vehicles')
plt.xlabel('Drive wheels');
```

# Number of doors bar chart

```python
auto_df['num-of-doors'].value_counts().plot(kind='bar',color='purple')
plt.title("Number of doors frequency diagram")
plt.ylabel('Number of vehicles')
plt.xlabel('Number of doors');
```

# CORRELATION ANALYSIS

```python
import seaborn as sns
corr = auto_df.corr()
sns.set_context("notebook", font_scale=1.0, rc={"lines.linewidth": 2.5})
plt.figure(figsize=(13,7))
a = sns.heatmap(corr, annot=True, fmt='.2f')
rotx = a.set_xticklabels(a.get_xticklabels(), rotation=90)
roty = a.set_yticklabels(a.get_yticklabels(), rotation=30)
```

# Bivariate Analysis

```python
#Boxplot of Price and make

plt.rcParams['figure.figsize']=(23,10)
ax = sns.boxplot(x="make", y="price", data=auto_df)
```

```python
#Scatter plot of price and engine size

g = sns.lmplot('price',"engine-size", auto_df)
```

```python
#Scatter plot of normalized losses and symboling

g = sns.lmplot('normalized-losses',"symboling", auto_df)
```

```python
#Scatter plot of Engine size and Peak RPM

plt.scatter(auto_df['engine-size'],auto_df['peak-rpm'])
plt.xlabel('Engine size')
plt.ylabel('Peak RPM')
```

```python
#Scatter plot of City and Highway MPG, Curb weight based on Make of the car

g = sns.lmplot('city-mpg',"curb-weight", auto_df, hue="make", fit_reg=False)
```

```python
g = sns.lmplot('highway-mpg',"curb-weight", auto_df, hue="make",fit_reg=False)
```

# Drive wheels and City MPG bar chart

```python
auto_df.groupby('drive-wheels')['city-mpg'].mean().plot(kind='bar', color = 'peru');
plt.title("Drive wheels City MPG")
plt.ylabel('City MPG')
plt.xlabel('Drive wheels')
```

# Drive wheels and Highway MPG bar chart


```python
auto_df.groupby('drive-wheels')['highway-mpg'].mean().plot(kind='bar', color = 'peru');
plt.title("Drive wheels Highway MPG")
plt.ylabel('Highway MPG')
plt.xlabel('Drive wheels')
```

# Normalized losses based on body style and no. of doors

```python
pd.pivot_table(auto_df,index=['body-style','num-of-doors'], values='normalized-losses').plot(kind='bar',color='purple')
plt.title("Normalized losses based on body style and no. of doors")
plt.ylabel('Normalized losses')
plt.xlabel('Body style and No. of doors')
```
