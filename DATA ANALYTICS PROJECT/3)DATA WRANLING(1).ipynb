```python
#READ DATA

import pandas as pd
import numpy as np
auto_df=pd.read_excel('C:/Users/tech2k8/data clean nxt to data wrangling.xlsx','auto')
auto_df.head(5) 
```

```python
#DATA WRANGLING

#GROUP BY

group=auto_df.groupby('make')
group
```

```python
#ITERATING MAKE COLUMN 
#SPLIT TO COMBINE

for make,a in group:
    print(make)
    print(a)
    
```

```python
#BMW CAR GROUP

group.get_group('bmw')
```

```python
#DESCRIBE()

print(group.describe())
```

```python
#PLOT

print(group.plot())
```

```python
#aggregation

group.agg(['max','min','std','mean'])
```
